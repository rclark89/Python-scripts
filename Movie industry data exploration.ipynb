```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts config of plots to be created.

#Read in the data.
#Create a dataframe.

df = pd.read_csv(r'C:\Users\Ross\Documents\py_port\movies.csv')
```

```python
#Let's check out the data.

df.head()
```

```python
#Check for missing data.
#Create for loop to loop through dataset and check for missing data.
#If every column is present, it will print a percentrage value for the number
#of empty 'cells' in that column.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing*100))
```

```python
#Create correct year column
df['yearcorrect'] = df['year'].astype(str).str[:4]
```

```python
#Sort data frame by 'gross' in descending order.
df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
#Scatter plot budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget per film')
plt.show()
```

```python
#Carry out regression to determine degree of correlation between variables.
#Using Seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.title('Budget vs. Gross Earnings')
plt.show()
```

```python
#Quantify correlation in numerical fields.
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
#Produce correlation matrix for dataframe.
correlation_matrix = df.corr(method='pearson')

#Visualise correlation matrix using Seaborn.
sns.heatmap(correlation_matrix, annot=True,)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
#'Company' is not a numeric category. However we can create a numeric representation.

df_numerical = df
#convert 'object' type (non-numerical) column data to 'category' (numerical) column data.
for col_name in df_numerical.columns:
    if(df_numerical[col_name].dtype == 'object'):
        df_numerical[col_name] = df_numerical[col_name].astype('category')
        #Assigns unique code to unique entries.
        df_numerical[col_name] = df_numerical[col_name].cat.codes

df_numerical         
            
    
```

```python
#Produce correlation matrix for dataframe using categorised data.
correlation_matrix = df_numerical.corr(method='pearson')

#Visualise correlation matrix using Seaborn.
sns.heatmap(correlation_matrix, annot=True,)
plt.title('Correlation Matrix for Numeric Features, inc all fields')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
